```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time

# Set up the driver (Make sure to replace 'chromedriver' with the path to your WebDriver)
driver = webdriver.Chrome()

# Open the APSRTC page on Redbus
driver.get('https://www.redbus.in')

# You might need to interact with the page to select APSRTC and a specific route.
# The following example assumes you're already on the correct APSRTC results page.

def extract_bus_info():
    buses = driver.find_elements(By.XPATH, '//div[contains(@class, "bus-item")]')
    for bus in buses:
        try:
            name = bus.find_element(By.XPATH, './/div[contains(@class, "travels")]').text
            departure_time = bus.find_element(By.XPATH, './/div[contains(@class, "dp-time")]').text
            arrival_time = bus.find_element(By.XPATH, './/div[contains(@class, "bp-time")]').text
            price = bus.find_element(By.XPATH, './/div[contains(@class, "seat-fare")]//span').text
            print(f"Name: {name}, Departure: {departure_time}, Arrival: {arrival_time}, Price: {price}")
        except Exception as e:
            print(f"Error extracting bus info: {e}")

def go_to_next_page():
    try:
        next_button = WebDriverWait(driver, 10).until(
            EC.element_to_be_clickable((By.XPATH, '//a[@class="next"]'))
        )
        next_button.click()
        return True
    except Exception as e:
        print(f"Error navigating to next page: {e}")
        return False

# Extract info from the first page
extract_bus_info()

# Loop through the next pages and extract info
while True:
    if not go_to_next_page():
        break
    time.sleep(5)  # Wait for the next page to load
    extract_bus_info()

# Close the driver
driver.quit()
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import time
import sqlite3

# Initialize the WebDriver
driver = webdriver.Chrome()

# URL of the website to scrape
url = 'https://www.redbus.in'

# Function to scrape data
def scrape_data():
    driver.get(url)
    time.sleep(3)  # Allow time for the page to load

    # Sample scraping logic
    buses = driver.find_elements(By.CLASS_NAME, 'bus-class')

    bus_data = []
    for bus in buses:
        route_name = bus.find_element(By.CLASS_NAME, 'route-name').text
        route_link = bus.find_element(By.CLASS_NAME, 'route-link').get_attribute('href')
        busname = bus.find_element(By.CLASS_NAME, 'bus-name').text
        bustype = bus.find_element(By.CLASS_NAME, 'bus-type').text
        departing_time = bus.find_element(By.CLASS_NAME, 'departing-time').text
        duration = bus.find_element(By.CLASS_NAME, 'duration').text
        reaching_time = bus.find_element(By.CLASS_NAME, 'reaching-time').text
        star_rating = float(bus.find_element(By.CLASS_NAME, 'star-rating').text)
        price = float(bus.find_element(By.CLASS_NAME, 'price').text.replace('₹', '').replace(',', ''))

        bus_data.append((route_name, route_link, busname, bustype, departing_time, duration, reaching_time, star_rating, price))

    return bus_data

bus_data = scrape_data()
driver.quit()

# Connect to the SQL database
conn = sqlite3.connect('bus_data.db')
cursor = conn.cursor()

# Create the table
cursor.execute('''
CREATE TABLE IF NOT EXISTS bus_data (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    route_name TEXT,
    route_link TEXT,
    busname TEXT,
    bustype TEXT,
    departing_time TEXT,
    duration TEXT,
    reaching_time TEXT,
    star_rating REAL,
    price REAL
)
''')

# Insert data into the table
cursor.executemany('''
INSERT INTO bus_data (route_name, route_link, busname, bustype, departing_time, duration, reaching_time, star_rating, price)
VALUES (?, ?, ?, ?, ?, ?, ?, ?, ?)
''', bus_data)

# Commit and close the connection
conn.commit()
conn.close()
Streamlit Application
python
Copy code
import streamlit as st
import sqlite3
import pandas as pd

# Connect to the SQL database
conn = sqlite3.connect('bus_data.db')
cursor = conn.cursor()

# Query the data
cursor.execute('SELECT * FROM bus_data')
rows = cursor.fetchall()

# Convert to DataFrame
columns = ['id', 'route_name', 'route_link', 'busname', 'bustype', 'departing_time', 'duration', 'reaching_time', 'star_rating', 'price']
df = pd.DataFrame(rows, columns=columns)

# Streamlit application
st.title('Redbus Data Filtering and Analysis')

# Filters
route_name_filter = st.selectbox('Select Route Name', df['route_name'].unique())
bus_type_filter = st.selectbox('Select Bus Type', df['bustype'].unique())
price_filter = st.slider('Select Price Range', int(df['price'].min()), int(df['price'].max()))

# Filter the data
filtered_df = df[(df['route_name'] == route_name_filter) & 
                 (df['bustype'] == bus_type_filter) & 
                 (df['price'] <= price_filter)]

# Display the data
st.dataframe(filtered_df)

# Close the connection
conn.close()
```
